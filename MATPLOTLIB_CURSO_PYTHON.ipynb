# SUBPLOT

```python
# CRIANDO NÚMEROS ALEATÓRIOS PARA OS SUBPLOTS:

x = np.arange(1, 11)
y1 = np.random.randint(1, 400, 10)
y2 = np.random.randint(150, 500, 10)
y3 = np.random.randint(200, 600, 10)
```

```python
plt.figure(figsize = (20, 5))
plt.suptitle('Titulo da Figura', fontsize = 15)

plt.subplot(1,3,1)  # 1 linha, 3 colunas, coluna 1
plt.plot(x, y1, color = 'k')
plt.title('Subplot 1', pad = 10)
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')

plt.subplot(1,3,2) # 1 linha, 3 colunas, coluna 2
plt.plot(x, y2, color = 'r')
plt.title('Subplot 2', pad = 10)
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')

plt.subplot(1,3,3) # 1 linha, 3 colunas, coluna 3
plt.plot(x, y3, color = 'g')
plt.title('Subplot 3', pad = 10)
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')


plt.tight_layout(pad=2)
plt.show()
```

```python
# OUTRO MODO DE FAZER SUBPLOT:

fig, ax = plt.subplots(1, 3, figsize = (20,5))
fig.suptitle('Figure')
ax[0].plot(x, y1, color = 'black')
ax[1].plot(x, y2, color = 'red')
ax[2].plot(x, y3, color = 'green')

for i in range(3):
  ax[i].set(title = f'Subplot {i+1}', xlabel = 'Eixo X', ylabel = 'Eixo Y')
```

**IMPORTANDO DADOS ECONÔMICOS DE UM DATASET DISPONIBILIZADO PELO PLOTLY.EXPRESS**

```python
import plotly.express as px
```

```python
px.data.gapminder()
```

```python
# armazenado os dados em um dataframe, filtrado apenas pelo Brasil como país e colocado como índice o ano

df = px.data.gapminder().query('country == "Brazil"').set_index('year')
```

```python
df.head()
```

```python
# GRÁFICO PARA VERIFICAR A EVOLUÇÃO DA RENDA PER CAPITA DO BRASIL:

plt.plot(df.index, df['gdpPercap'])  # df.index = tempo/ano
plt.title('PIB per Capita do Brasil')
plt.ylabel('PIB per Capita')
plt.xlabel('Tempo')
plt.tight_layout(pad=2)
```

```python
# GRÁFICO PARA VERIFICAR A RELAÇÃO ENTRE A ESPECTATIVA DE VIDA E A RENDA PER CAPITA NO BRASIL:

plt.figure(figsize = (12, 4))
plt.scatter(df['lifeExp'], df['gdpPercap'], cmap = 'viridis')
plt.title('Relação entre a expectativa de vida e renda per Capita no Brasil', loc = 'left')
plt.ylabel('Renda per Capita')
plt.xlabel('Expectativa de Vida')
plt.tight_layout(pad=2)
plt.show()
```

# GRÁFICO DE BARRAS

```python
plt.bar(x = df.index,
        height = df['pop'],
        color = 'red'
        )

plt.title('População Brasileira')
```

```python
# FAZER FUNÇÕES PARA FILTRAR AMOSTRAS NO DATASET:

def filtrar_continente(continente):
  df = px.data.gapminder()
  df = df[df['continent'] == continente]
  return df


def filtrar_pais(pais, variavel):
  'Filtrar algum país da amostra para as variáveis "pop", "gdpPercap" e "lifeExp"'
  df = px.data.gapminder()
  df = df[df['country'] == pais][variavel]
  return df
```

```python
americas = filtrar_continente('Americas')
paises = americas['country'].unique()
plt.figure(figsize = (12, 8))

for pais in paises:
  plt.scatter(filtrar_pais(pais = pais,
                           variavel = 'lifeExp'),
                           filtrar_pais(pais = pais,
                                        variavel = 'gdpPercap'))

plt.legend(labels=paises, loc = 'best')
plt.title('Relação entre renda per capita e expectativa de vida', loc = 'left')
plt.xlabel('Expectativa de Vida')
plt.xlabel('Renda per Capita');
```
